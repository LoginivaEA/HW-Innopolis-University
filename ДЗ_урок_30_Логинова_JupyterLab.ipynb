```python
Задача 1. Фильтрация и трансформация данных (lambda, map, filter)
1. Используя filter и lambda, отфильтруйте студентов, чья оценка (grade) выше 80.
2. Используя map и lambda, преобразуйте отфильтрованный список, оставив только имена студентов в верхнем регистре.
3. Используя reduce (из functools) и lambda, вычислите среднюю оценку всех студентов из исходного списка. Результат округлите до двух знаков после запятой.
```

```python
from functools import reduce

students = [
    {'name': 'Алексей', 'grade': 85, 'city': 'Москва'},
    {'name': 'Мария', 'grade': 92, 'city': 'Казань'},
    {'name': 'Иван', 'grade': 78, 'city': 'Москва'},
    {'name': 'Елена', 'grade': 88, 'city': 'Иннополис'},
    {'name': 'Дмитрий', 'grade': 55, 'city': 'Москва'},
]

# 1. filter
filtered = list(filter(lambda s: s['grade'] > 80, students))

print("Отфильтрованные студенты:")
for s in filtered:
    print(s)

# 2. map
names = list(map(lambda s: s['name'].upper(), filtered))

print("\nИмена в верхнем регистре:")
print(names)

# 3. reduce (средняя оценка)
avg = reduce(lambda acc, x: acc + x['grade'], students, 0) / len(students)

print("\nСредняя оценка:", round(avg, 2))
```

```python
Задача 3. Вложенные функции
Напишите программу, которая выполняет следующие действия:
1. Определите функцию get_list, которая принимает строку из целых чисел, разделенных пробелами, и возвращает список из этих чисел.
2. Создайте функцию sort_func, которая:
a. Принимает функцию как аргумент
b. Вызывает переданную функцию
c. Сортирует результат работы функции по возрастанию
d. Возвращает отсортированный результат
3. Используйте вызов функций get_list и sort_func, чтобы вывести результат на экран.
Выход
```

```python
# 1. get_list
def get_list(string):
    return list(map(int, string.split()))

# 2. sort_func
def sort_func(func):
    result = func()
    return sorted(result)

# 3. использование функций
input_string = "3 1 4 1 5 9 2 6"

def wrapper():
    return get_list(input_string)

print("Исходный список:", wrapper())
print("Отсортированный:", sort_func(wrapper))
```

```python
Задача 4. Декоратор для измерения времени выполнения
Реализуйте декоратор timer, который:
1. Замеряет время выполнения декорируемой функции
2. Выводит на экран сообщение: "Функция {имя_функции} выполнилась за {время:.4f} сек."
3. Возвращает результат выполнения функции
4. Корректно работает с любыми функциями (любое количество аргументов)
5. Сохраняет имя (__name__) и документацию (__doc__) исходной функции
```

```python
# 1. Импорт
import time
from functools import wraps

# 2. Декоратор timer
def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        
        result = func(*args, **kwargs)
        
        end = time.time()
        
        print(f"Функция {func.__name__} выполнилась за {end - start:.4f} сек.")
        
        return result
    
    return wrapper
    
# 3. Функция с суммой
@timer
def sum_to_n():
    """Сумма чисел от 1 до 1_000_000"""
    return sum(range(1, 1_000_001))

# 4. Запуск
result = sum_to_n()

print("Результат:", result)
```
