### Задача 1: Чтение и печать данных в консоли

**Описание:** Программа считывает разделитель, три строки, символ завершения и одну строку,
затем выводит указанные значения через разделитель с учетом символа завершения.

```python
sep = input("Введите строку-разделитель (sep): ")

first = input("Введите первое значение (first column): ")
second = input("Введите второе значение (second column): ")
third = input("Введите третье значение (third column): ")

end = input("Введите символ завершения строки (end): ")

other = input("Введите другие данные: ")

print(first, second, third, other, sep=sep, end=end)
```

### Задача 2: Максимум, минимум, остальное

**Описание:** Программа получает три целых числа и выводит сначала максимальное,
потом минимальное, затем оставшееся число. Решение без использования функций min() и max().

```python
a = int(input())
b = int(input())
c = int(input())

# Находим максимум
if a >= b and a >= c:
    maximum = a
elif b >= a and b >= c:
    maximum = b
else:
    maximum = c

# Находим минимум
if a <= b and a <= c:
    minimum = a
elif b <= a and b <= c:
    minimum =b
else:
    minimum = c


print(maximum, minimum, a + b + c - maximum - minimum, sep="\n")
```

### Задача 3: Простой калькулятор

**Описание:** Программа считывает два вещественных числа и операцию,
затем применяет операцию к числам и выводит результат.
Поддерживаемые операции: +, -, /, *, mod, pow, div.
При делении на 0 выводит "Деление на 0!".

```python
# Считываем два вещественных числа
a = float(input())
b = float(input())

# Считываем операцию
operation = input()

# Проверяем деление на 0 для операций деления
if (operation == "/" or operation == "div" or operation == "//" or
    operation == "mod" or operation == "%"):
    if b == 0.0:
        print("Деление на 0!")
    else:
        if operation == "/":
            print(a / b)
        elif (operation == "/" or operation == "div"):
            print(a // b)
        elif (operation == "mod" or operation == "%"):
            print(a % b)

else:
    # Операции, не требующие проверки на 0
    if operation == "+":
        print(a + b)
    elif operation == "-":
        print(a - b)
    elif operation == "*":
        print(a * b)
    elif (operation == "**" or operation == "pow"):
        print(a ** b)
```
